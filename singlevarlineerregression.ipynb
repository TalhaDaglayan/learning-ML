```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import os
```

```python
df = pd.read_excel(os.path.join('.', 'data', 'houseprices.xlsx'))
df
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('price ($)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
area = df[['area']].values.reshape(-1, 1)
reg.fit(area, df.price)

```

```python
X_pred = np.array(3300).reshape(-1, 1)
reg.predict(X_pred)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('price ($)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
