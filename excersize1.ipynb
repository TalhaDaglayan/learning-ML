```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import os
```

```python
df = pd.read_csv(os.path.join('.', 'data', 'canada_per_capita_income.csv'))
df.rename(columns={'per capita income (US$)': 'percapitaincome'}, inplace=True)
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income per capita ($)')
plt.scatter(df['year'], df['percapitaincome'], color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
year = df[['year']].values.reshape(-1, 1)
reg.fit(year,df.percapitaincome)
```

```python
X_pred = np.array(2020).reshape(-1, 1)
reg.predict(X_pred)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income per capita ($)')
plt.scatter(df['year'], df['percapitaincome'], color='red', marker='+')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
```
